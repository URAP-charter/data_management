```python
# Initialize text strings and counts as empty, then convert data types:
empty = ["" for elem in range(len(df_charters["NCESSCH"]))] # Create empty string column that is as long as the longest variable (NCESCCH used for matching)
df_charters = df_charters.assign(SEARCH16=empty, TOTETH=empty, PCTETH=empty, AGE=empty, PCTFRL=empty, PLACE=empty) # Add empty columns to df
df_charters.loc[:,["PLACE", "SEARCH16"]] = df_charters.loc[:,["PLACE", "SEARCH16"]].apply(lambda x: x.astype(object)) # Convert to object type--holds text
df_charters.loc[:,["AGE", "TOTETH"]] = df_charters.loc[:,["AGE", "TOTETH"]].apply(pd.to_numeric, downcast="unsigned") # Convert to int dtype--holds positive numbers (no decimals)
df_charters.loc[:,["PCTETH", "PCTFRL"]] = df_charters.loc[:,["PCTETH", "PCTFRL"]].apply(pd.to_numeric, downcast="float") # Use most efficient float type for these vars--hold decimals
```

```python
data_year = int(2015)

# Compute demographic variables:
df_charters["TOTETH"] = df_charters[["AM", "AS", "BL", "HI", "HP", "TR"]].apply(sum, axis=1) # Number of nonwhite K-12 students
df_charters["PCTETH"] = (df_charters["TOTETH"]/df_charters["MEMBER"]).apply(pd.to_numeric, downcast='float')  # Percent nonwhite K-12 students
df_charters["PCTFRL"] = (df_charters["TOTFRL"]/df_charters["MEMBER"]).apply(pd.to_numeric, downcast='float')  # Percent receiving free/ reduced-price lunch
df_charters["AGE"] = data_year - df_charters["YEAR_OPENED"] # Number of years school has been open

df_charters["SEARCH16"] = df_charters["SCHNAM16"].apply(lambda x: str(x) + " ") + df_charters["ADDRESS16"]

# Recode variables:
df_charters["PLACE"] = df_charters["LOCALE15"].map({11.0:"City", 12.0:"City", 13.0:"City", 21.0:"Suburb", 22.0:"Suburb", 23.0:"Suburb", 31.0:"Town", 32.0:"Town", 33.0:"Town", 41.0:"Rural", 42.0:"Rural", 43.0:"Rural"}).astype('category')
df_charters["LOCALE15"] = df_charters["LOCALE15"].map({11.0:"City (large)", 12.0:"City (midsize)", 13.0:"City (small)", 21.0:"Suburb (large)", 22.0:"Suburb (midsize)", 23.0:"Suburb (small)", 31.0:"Town (fringe)", 32.0:"Town (distant)", 33.0:"Town (remote)", 41.0:"Rural (fringe)", 42.0:"Rural (distant)", 43.0:"Rural (remote)"}).astype('category')
df_charters["TITLEI"] = df_charters["TITLEI"].map({"Yes":1, "No":0}).astype('category')
```
