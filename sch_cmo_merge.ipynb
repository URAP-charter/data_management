```python
import pandas as pd
#assumes that both data files and script are in same file

df_cmo = pd.read_excel("CMOs 2014-15_062618.xlsx", encoding = "latin1")
df_csch = pd.read_csv("pub_schools_merged_2015.csv", encoding = "latin1")
```

```python
list(df_csch)
```

```python
# goal: to reshape CMO dataframe so that it can be joined with  the charter school data on SCH_NAME

# reshapes the CMO dataframe by setting index to CMO_NAME, filtering out the columns with SCH_NAME in name, and then stacks SCH_NAMES with CMO_NAME
df_cmod = df_cmo.set_index('CMO_NAME',inplace = False)
df_cmod = df_cmod.filter(regex=("SCH_NAME.*"))
df_cmod = df_cmod.stack()
df_cmod
```

```python
# turns CMO_NAME to being a column instead of index
# result: CMO dataframe has school names grouped with CMO_NAME

df_cmod = df_cmod.reset_index(level = ['CMO_NAME'])
df_cmod.columns = ['CMO_NAME', 'SCH_NAME']
df_cmod
```

```python
# projects cmo_data onto required columns and then renames columns accordingly

column_list = ['SCH_NAME','CMO_NAME','STATE_MANUAL','NUMSCHOOLS_MANUAL17','URL','NUM_STATES']
df_newcmo = pd.merge(df_cmod,df_cmo, how = 'left', on = ['CMO_NAME'])[column_list]
df_newcmo.rename(columns={'STATE_MANUAL':'CMO_STATE','NUMSCHOOLS_MANUAL17':'CMO_SCHNUM','URL':'CMO_URL','NUM_STATES':'CMO_NUMSTATES'}, inplace=True)
df_newcmo['SCH_NAME'] = df_newcmo['SCH_NAME'].str.upper()
```

```python
df_newcmo
```

```python
df_csch.shape
```

```python
# performs merge with charter school data on 'SCH_NAME' and stores into 'result'
# groups data by CMO_NAME and finds sum of 'member' column
# merges/appends sum column with result
df_csch['NAME'] = df_csch['NAME'].str.upper()
result = pd.merge(df_newcmo, df_csch, how = 'right', right_on = ['NAME'], left_on = ['SCH_NAME'])
mem_col = result.groupby('CMO_NAME').sum()
mem_col = mem_col.reset_index(level = ['CMO_NAME'])
mem_col = mem_col[['CMO_NAME', 'POPTOT_S16']]
mem_col.columns = ['CMO_NAME', 'CMO_MEMSUM']
result = pd.merge(mem_col, result, how = 'right', on = ['CMO_NAME'])
```
