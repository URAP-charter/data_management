```python
#Uploading most recent csv file of charter school data
charter_schools =pd.read_csv('../../nowdata/parsing/overlaps_removed_df.csv', sep = '\t', header = 0)
```

```python
#Selecting the columns that are needed to calculate the ethnic concentrations 
#WH = Percentage of Students - White
# AS= Percentage of Students - Asian 
# BL= Percentage of Students - Black 
# HI = Percentage of Students - Hispanic 
charter_schools[[ 'WH','AS','BL','HI']]
 
```

```python
#Goal: Add a column that tells whether the school has an Asian population greater than 98% 
asian_98= charter_schools['AS']>= 98
# Convert to list because the a series will not append to a dataframe 
asian_98=asian_98.tolist()
charter_schools['98%_ASIAN'] = asian_98




```

```python
# Number of schools that have an Asian population greater than 98%
total_schools_with_asian_pop_greater_than_98 = np.count_nonzero(charter_schools['AS']>= 98)
total_schools_with_asian_pop_greater_than_98
```

```python
#Goal: Add a column that tells whether the school has Black population greater than 98% 
black_98= charter_schools['BL']>= 98
black_98 =black_98.tolist()
charter_schools['98%_BLACK'] = black_98
```

```python
# Number of schools that have a Black student population greater than 98%
total_schools_with_black_pop_greater_than_98 = np.count_nonzero(charter_schools['BL']>= 98)
total_schools_with_black_pop_greater_than_98
```

```python
#Goal: Add a column that tells whether the school has a White student population greater than 98% 
WH_98= charter_schools['WH']>= 98
WH_98=WH_98.tolist()
charter_schools['98%_WHITE'] = WH_98

```

```python
# Number of schools that have a White population greater than 98%
total_schools_with_white_pop_greater_than_98 = np.count_nonzero(charter_schools['WH']>= 98)
total_schools_with_white_pop_greater_than_98
```

```python
#Goal: Add a column that tells whether the school has a Hispanic population that is greater than 98% 
HI_98= charter_schools['HI']>= 98
HI_98=HI_98.tolist()
charter_schools['98%_HISPANIC'] = HI_98

```

```python
# Number of schools that have a Hispanic population greater than 98%
total_schools_with_hispanic_pop_greater_than_98 = np.count_nonzero(charter_schools['HI']>= 98)
total_schools_with_hispanic_pop_greater_than_98
```

```python
# In order to find the schools that has a population of non-white students greater than 98% add the columns asian, black, and hispanic. 
# Next convert the series into a list in order to be able to append the non_white column to the data frame 
asian= charter_schools['AS']
black = charter_schools['BL']
HI = charter_schools['HI'] 
non_white = asian + black + HI
non_white_98 = non_white >= 98
non_white_98 = non_white_98.tolist()
charter_schools["98%_NONWHITE"] = non_white_98

   

```
