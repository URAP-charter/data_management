# Step 1: Stack Duplicates on top of eachother (sort_values technique)

Below we assigned column "NCESSCH" to key so that we can reference that column easily. We sorted the dataframe in a way that stacks rows with duplicate keys right on top of eachother so you can see each group of duplicate keys right next to eachother rather than scattered randomly throughout dataframe.

Helpful Links:

https://stackoverflow.com/questions/44123874/dataframe-object-has-no-attribute-sort


https://stackoverflow.com/questions/14657241/how-do-i-get-a-list-of-all-the-duplicate-items-using-pandas-in-python


```python
key = charterdf["NCESSCH"]
sorted_duplicates = charterdf[key.isin(key[key.duplicated()])].sort_values('NCESSCH')
sorted_duplicates 

```

```python
type(sorted_duplicates)
```

```python
sorted_duplicates["is_duplicate"]= sorted_duplicates.duplicated()
 
sorted_duplicates
```

```python
sorted_duplicates.duplicated(subset=None, keep='first')
```
