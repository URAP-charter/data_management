```python
% matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import gc
```

```python
path = '../../charters_full_2015.pkl'
gc.disable()
df_charter = pd.read_pickle(path)
gc.enable()
```

```python
df_charter[df_charter['SCHNAM16'] == 'TSU CHARTER LAB SCH']['WEBTEXT'] # webtext of school with 2k+ pages post filter. Its webtext
# was replaced by its CMO's webtext. Therefore the CMO had 2k+ pages with >1.5 hit score.
```

```python
df_charter['WEBTEXT']=df_charter['WEBTEXT'].fillna('')
```

```python
df_charter['NUMPAGES'] = df_charter['WEBTEXT'].apply(len)
```

```python
df_charter['NUMPAGES'].describe()
```

```python
df_charter['NUMPAGES'].sum()
```

```python
set(sorted(df_charter['NUMPAGES'].values))
```

```python
df_charter['NUMPAGES'].value_counts().sort_index() # one school has 17k pages
```

```python
list(set(sorted(df_charter['NUMPAGES'].values)))[150] # 150th lowest distinct number of pages. happens to be 150
```

# BAR GRAPHS: SET LIMITS

```python
df_charter['NUMPAGES'].value_counts().sort_index().iloc[:150] # counts for lowest 150 distinct number of pages
```

```python
df_charter['NUMPAGES'][df_charter['NUMPAGES']>1][df_charter['NUMPAGES']<250].values.sum()
```

```python
df_charter['NUMPAGES'].hist(bins = 100) # whole distribution
plt.show()
```

```python
barlow = 4
barhigh = 250 # set this variable to be the max number of pages you want to see the counts for in the final histogram
```

```python
pagenum_window = df_charter['NUMPAGES'][barhigh>=df_charter['NUMPAGES']][barlow<=df_charter['NUMPAGES']] # eliminates upper vaues to get a better look at distribution for lower page numbers
```

```python
pagenum_window.hist(bins = 50) # better look using the limit variables set above
plt.show()
```
